Here a particular dataset regarding iris is obtained.At first the modules are imported and then tha data is loaded by reading   the data.Then columns,mean and null values of provided data is discovered.By using histogram,scatterplot,pairplot,stripplot and  boxplot a visualization of data is established.

# import modules

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

# Loading the Dataset

```python
dataset=pd.read_excel("iris.xls")
```

# Visualization

# 1.Histogram

Distribution of SL in histogram

```python
plt.figure()
plt.hist(dataset['SL'],rwidth=0.9)
plt.title('Distribution of SL',fontsize=16)
plt.xticks()
plt.yticks()
```

Distribution of PL in histogram

```python
dataset['PL'].hist()
plt.title('Distribution of PL')
```

Distribution of SW in histogram

```python
dataset['SW'].hist()
plt.title('Distribution of SW')
```

Distribution of PW in histogram

```python
dataset['PW'].hist()
plt.title('Distribution of PW')
```

# 2.Scatter Plot

Distribution of SL vs SW in scatterplot

```python
plt.figure()
plt.scatter(dataset['SL'],dataset['SW'],s=10,c='red')
plt.title('Distribution of SL vs SW',fontsize=16)
plt.xlabel('Sepal Length')
plt.ylabel('Sepal Width')
plt.xticks()
plt.yticks()
```

Distribution of PL vs PW in scatterplot

```python
plt.figure()
plt.scatter(dataset['PL'],dataset['PW'],s=10,c='red')
plt.title('Distribution of PL vs PW',fontsize=16)
plt.xlabel('Petal Length')
plt.ylabel('Petal Width')
plt.xticks()
plt.yticks()
```

# 3.Pair Plot

```python
sns.pairplot(dataset)
```

# 4.Strip Plot

Stripplot distribution between SL and SW

```python
sns.stripplot(x='SL',y='SW',data=dataset)
```

Stripplot distribution between PL and PW

```python
sns.stripplot(x='PL',y='PW',data=dataset)
```

# 5.Box Plot

Boxplot distribution between SL and SW

```python
sns.boxplot(x='SL',y='SW',data=dataset)
```

Boxplot distribution between PL and PW

```python
sns.boxplot(x='PL',y='PW',data=dataset)
```
